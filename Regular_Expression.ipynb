```python
import re
```

```python
txt = "Today  in fifa  world cup Brazil won "
x = re.findall(r"[a-z]",txt)
print(x)
```

***when we use this [a-z] it only find out the small word in a to z .not find out the capital word we see that "T" and 'B' is not print here ***

```python
txt = "Today is a beautiful rainy day "
x = re.findall(r"[aeiou]",txt) # find out all vowel from the text by using this
x
```

```python
txt2 = "My phone Number is 013"

x = re.findall(r"[a-zA-Z0-9]",txt2)
x# select aall the capital and small word and also number in details
```

# Only number show

```python
# find all digits charecters
x = re.findall(r"[\d]",txt2)
x
# select only the digit from the text

```

# Digit charecter class

```python
txt = "That will be 59 dollers"
x = re.findall(r"\d",txt)
# find the digit charecter
x
```

```python
txt = "That will be 59 and 78 899 dollers"
x = re.findall(r"\d+",txt)
# find the digit charecter
x
```

```python
# match 2 , 3 digits

txt = "That will be 59 and 78 899 dollers"
x = re.findall(r"\d{2}",txt)
# find the digit charecter
x
```

```python
# 2 or 3 or more  digit selection
txt = "That will be 59 and 78 899 dollers"
x = re.findall(r"\d{2,3}",txt)
# find the digit charecter
x
```

```python
num = "01302-880503"
x = re.findall(r"\d{5,6}",num)# 5 and 6 digit selection
x
```

```python
num = "01302-880503"
x = re.findall(r"\d{5}-\d{6}",num)
x
```

```python
phn="+8801302880503"
x = re.findall(r"\+\d{13}",phn)
x
```

```python
# start at the beginning

txt = "123 123 123 this is my phone number 01302880503"
x = re.findall(r"^\d+",txt) # ^ dile first ar selected jinish print kore dibe

x
```

```python
# At the end print

txt = "123 123 123 this is my phone number 01302880503"
x = re.findall(r"\d+$",txt)# last a number ase kin check korese jodi  number na thakto tahole error show korto
x
```
