# BAg of words

```python
import numpy as np
import pandas as pd
```

```python
dict ={'text':['peopole watch youtube','youtube watch youtube','people write comment','youtube write comment'],'outputs':[0,0,1,1]}
```

```python
df=pd.DataFrame(dict)
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bag_of_words = cv.fit_transform(df['text'])
```

```python
bag_of_words
```

```python
#Vocab

cv.vocabulary_# number is this given is index number
```

```python
print(bag_of_words[0].toarray())
```

```python
print(bag_of_words[1].toarray())
```

এইখানে আমরা দেখতে পেলাম যে One Hot Encoding-এ আমাদের যে সমস্যা ছিল, সেই সমস্যা এখানে সমাধান হলো। যেমন, one-hot encoding এর মেইন সমস্যা ছিল যে, আমাদের যদি কোন word এর one-hot encoding করা না থাকতো, তাহলে আমাদের model সেই word টাকে চিনতে পারত না। কিন্তু "Words of Bag"-এ এই সমস্যা-টি নেই। এখানে আমরা দেখতে পেলাম যে "and" এবং "of" নতুন একটি শব্দ, কিন্তু তারপরও আমাদের মডেলটি "and" এবং "of" কে ইনপুট হিসেবে নিচ্ছে এবং কোনো সমস্যা সৃষ্টি করছে না।

কিন্তু, ওয়ান হট এনকোডিংয়ে যদি আমাদের মডেলটি নির্দিষ্ট ভোকাবুলারির দ্বারা ট্রেনিং না হয় এবং নতুন কোনো শব্দ পায়, তাহলে মূলত সে এরর দিত। কিন্তু এখানে কোনো এরর দিচ্ছে না।
