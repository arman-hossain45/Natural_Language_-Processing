```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
story  = []
for filename in os.listdir('data'):
  f = open(os.path.join('data',filename))
  corpus = f.read()
  raw_sent = sent_tokenize(corpus)
  for sent in raw_sent:
    story.append(simple_preprocess(sent))
```

```python
model = gensim.models.Word2Vec(
    window = 10,
    min_count = 2
)
```

```python
model.build_vocab(story)# unique sword niye vocab create korbe
```

Gensim Word2Vec model-এর সবচেয়ে গুরুত্বপূর্ণ অংশগুলোর একটি। এটি model-কে আপনার dataset (story) থেকে শব্দগুলোর vector representation শেখাতে (train করতে) ব্যবহৃত হয়।

```python
model.corpus_count# story te koto gula sentence ase ta count korbe
```

```python
# train the hole model
model.train(story, total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.epochs# default  epochs is 5 in gensim library
```

```python
model.wv.most_similar('daenerys')# daenerys word ar sathe joto gula simillar word ase ta print korbe
```
