```python
exclude = string.punctuation
```

```python
def remove_pun(text):
  for char in exclude:
    text = text.replace(char,"")

  return text

remove_pun("hi, my name is arman hossain, and i was a brilliant student at a time ")
```

**way 2**

more efficient way it is

```python
def remove_pun1(text):
  return text.translate(str.maketrans(' ',' ',exclude))


remove_pun1("hi, my name is arman hossain, and i was a brilliant student at a time ")
```

# Stopwords remove

Stop Words কী?

Stop Words হলো এমন কিছু সাধারণ শব্দ, যেগুলো একটি বাক্যে খুব বেশি ব্যবহৃত হয় কিন্তু সাধারণত বাক্যের মূল অর্থ বোঝার জন্য খুব বেশি গুরুত্বপূর্ণ নয়।

উদাহরণ (English):

I am going to school

এখানে:

I
am
to

এগুলো Stop Words হতে পারে।

আর

going
school

এগুলো গুরুত্বপূর্ণ শব্দ।

```python
import nltk
```

```python
from nltk.corpus import stopwords

nltk.download('stopwords')
stopwords.words('english')
```

```python
stops_words = set(stopwords.words('english'))

def remove_stopwords(text):
  words = text.split()
  cleaned = []
  for word in words:
    # word convert in lower text
    word = word.lower().strip(string.punctuation)
    #punctuation thakle tao remove kore dibe aikhan a
    if word and word not in stops_words:
      cleaned.append(word)

  return " ".join(cleaned)
```

```python
remove_stopwords('i am going, to school')
```

```python
df['review']=df['review'].apply(remove_stopwords)
df['review']
```

# Remove Emoji  in text

```python
text_With_emoji ="""Hey! 😊 How are you doing today? 🌞
I'm feeling great 😄 and ready to learn some Python 🐍💻."""
```

```python
import re

def remove_emoji(text):
  emoji_pattern = re.compile("["# emoji unicode range
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               "]+",
                              flags=re.UNICODE)
  return emoji_pattern.sub(r'',text)
```

```python
clean_text = remove_emoji(text_With_emoji)
clean_text
```

```python
df['review'] = df['review'].apply(remove_emoji)
df['review']
```

# Stemming

simple word in bangla bhaga word ke bhagha

Stemming কী?

Stemming হলো NLP (Natural Language Processing)-এর একটি Text Preprocessing Technique, যেখানে একটি শব্দের বিভিন্ন রূপ (word forms) থেকে মূল অংশ (root/stem) বের করা হয়।

সহজ ভাষায়:

একটি শব্দের শেষে থাকা prefix/suffix কেটে শব্দটিকে তার মূল রূপের কাছাকাছি নিয়ে আসার প্রক্রিয়াকে Stemming বলে।

কেন Stemming দরকার?

ধরো নিচের শব্দগুলো আছে:

play
playing
played
plays

মানুষ বুঝতে পারে এগুলো একই শব্দ "play" থেকে এসেছে।

কিন্তু কম্পিউটার এগুলোকে ৪টি আলাদা শব্দ হিসেবে দেখে।

Stemming করলে:

play
playing  → play
played   → play
plays    → play

এখন সবগুলোকে একই শব্দ হিসেবে গণ্য করা যাবে।



```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = 'walk walks walking walked'
stem_words(sample)
```

```python
text = 'Probably my alltime favourite movie a story selflessness sacrifice and dedication to  be a good man and a good researcher'
stem_words(text)
```
